```python
pf=pf.apply(lambda x:x.str.strip().str.title() if x.dtype=='object' else x)
pf
```

**Data cleaning fix inconsistent text** lambda x: apply to each column(if x.dtype == "object") this check column is text or str data if it is not text it skip it(x.str.strip) remove unwanted space(.str.title)fist letter uppers case rest lower(else) if numeric it keeps unchage

```python
def extract_score(x):
    if pd.isna(x):
        return np.nan

    try:
        return int(str(x).split()[0])
    except:
        return np.nan

for col in ['sexual_activity', 'concentration', 'optimisim']:
    pf[col] = pf[col].apply(extract_score)
pf
```

{if pd.isna(x)}  check value x is missing (nan) if value miss (return np.nan) it keeps cell as nan try and except =it tries to extract a numeric score from string str()=make sure the value is string (.split())=split the string into parts by spaces eg='3 from 10'='3' 'from''10' ==[0] take the first which is 3

 for col in ['try-explanation','concentration','optimisim']:= this loops goes 3 column in data frame


```python
freq_map={
    'NO':0,'Seldom':1,'Somtime':2,'Usually':3,'Most_Often':4
}
freq_cols = ['sadness','euphoric','exhausted','sleep_dissorder']
for c in freq_cols:
    pf[c+'_ord']=pf[c].map(freq_map)
pf
```

**frequency column**c+'_ord' creates a new column map(freq_map) replaces the text values in column c with the corresponding numbers from

```python
yes_no_map = {'Yes':1,'No':0}
binary_cols = ['mood_swing','suicidal_thoughts','anorxia','authority_respect',
               'aggressive_response','ignore_&_move-on','nervous_break-down',
               'admit_mistakes','overthinking']
for c in binary_cols:
  pf[c+'_bin'] = pf[c].map(yes_no_map)
pf

```
